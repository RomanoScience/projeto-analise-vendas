```python
import pandas as pd
import matplotlib.pyplot as plt
#carregar o arquivos de dados
df = pd.read_csv("dados/vendas.csv.txt", sep = ",") 
# cmd para mostrar as primeiras linhas da tabela
df.head() 
media = df["Total"].mean()
print(f" valor medio de cada venda foi de R$:{media:.2f}") 
soma = df["Total"].sum()
print(f" Total vendido no periodo é R$:{soma}")
# total vendido por data
tot_data = df.groupby("Data")["Total"].sum()
print("\ntotal vendido por data:")
print(tot_data)
# Contagem da vendas por produto
cont_venda = df.groupby("Produto").count()
print(" \nContagem de vendas por produto :")
print(cont_venda)
#  PLOTAGEM DOS DADOS PARA VIZUALIZAR MELHOR    
tot_data.plot(kind="area",color = "skyblue")
plt.xlabel(tot_data)
plt.ylabel(soma)
plt.title("Total vendido por data")
plt.grid()
plt.tight_layout()
plt.show()
```
